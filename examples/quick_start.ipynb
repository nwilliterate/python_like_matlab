## 1. Array Creation

```python
A = zeros(3, 4)
B = ones(2, 3)
C = eye(3)
D = rand(3, 3)

print("D = rand(3, 3)")
print(D)
```

## 2. Plotting Graphs

```python
x = linspace(0, 2*3.14159, 100)
y = sin(x)

figure(figsize=(10, 4))
plot(x, y, 'b-', linewidth=2)
xlabel('x')
ylabel('sin(x)')
title('Sine Wave')
grid('on')
show()
```

## 3. Matrix Operations

```python
A = rand(3, 3)
A_inv = inv(A)
result = dot(A, A_inv)

print("A * inv(A) =")
print(result)
```

## 4. Statistics

```python
data = randn(100, 1)

print(f"Mean: {mean(data):.4f}")
print(f"Standard Deviation: {std(data):.4f}")
print(f"Maximum: {max(data):.4f}")
print(f"Minimum: {min(data):.4f}")
```
